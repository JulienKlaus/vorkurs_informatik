### Aufgabe 4 
Berechnen Sie im binären Zahlensystem und diskutieren Sie das Distributivgesetz im  binären Zahlensystem.
1. $0101\cdot(1011+0011)$
2. $0101\cdot1011+0101\cdot0011$

### Lösung
1. Zuerst rechnen wir $1011+0011=1110$ aus und danach die Multiplikation $0101\cdot 1110=1110+111000=1000110$
2. Hier rechnen wir zuerst $0101\cdot 1011=1011+101100=110111$ und $0101\cdot 0011=0101+01010=1111$. Nun können wir diese beiden addieren und erhalten $110111+1111=1000110$. 

Hier sehen wir, dass beides mal das gleiche heraus kommt, wir nehmen an, dass das Distributivgesetz gilt, können dies natürlich nicht Aufgrund eines Beispieles folgern. 

### Aufgabe 5
Berechnen Sie die entsprechende dezimale Zahl. Überprüfen Sie es mit https://www.h-schmidt.net/FloatConverter/IEEE754de.html.
1. $11000000111100000000000000000000_2$
2. $11000010110000000000000000000000_2$
3. $01000011100001010110000000000000_2$

### Lösung
Um die Zahl umzurechnen schauen wir uns die drei Teile einmal getrennt an $1|10000001|11100000000000000000000_2$. Wir sehen am ersten Teil $1$, dass die Zahl negativ ist. Nun rechnen wir den 2. Teil in eine Dezimalzahl um $10000001_2=129$. Nun rechnen wir $129-127=2$. Um so viele Stellen müssen wir den dritten Teil nach rechts verschieben. Am Anfang des dritten Teiles steht zudem eine 1 Komma. Dies ist im Standart so festgelegt. Der dritte Teil mit der führenden zusätzlichen 1 und dem verschobenen Komma lautet dann $111.1_2$. Wir rechnen diese Zahl nun um und dürfen nicht vergessen, dass die Zahl negativ ist. Die Lösung ist somit $-7.5$.

Die beiden anderen Zahlen sind äquivalent umzurechnen. Die Lösungen lauten

2. -96
3. 266.75
