```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import data, io, color

# !pip install opencv-python
```

```python
file = "lena_noise.PNG"
image = color.rgb2gray(io.imread(file))

io.imshow(image)
```

```python
rows,cols = image.shape

denoised = np.zeros((rows, cols))

size = 3
boarder = int(np.floor(size/2))

# setze mich beliebig aber quadratisch
gauss_filter = np.ones((size,size)) / (size*size)
print(gauss_filter)

# falls beliebig setze ich den size neu
size = len(gauss_filter)

for i in range(boarder,rows-boarder):
    for j in range(boarder,cols-boarder):
        snipped = image[i-boarder:i+1+boarder, j-boarder:j+1+boarder]
        new_value = np.sum(np.multiply(snipped,gauss_filter))
        denoised[i,j] = new_value

fig, axes = plt.subplots(1, 2, figsize=(16, 8))
axes[0].imshow(image, cmap=plt.cm.gray)
axes[1].imshow(denoised, cmap=plt.cm.gray)
```
